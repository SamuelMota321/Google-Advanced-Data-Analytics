### Are in your dataset missing or incomplete data?
  - Decision: Drop the 700 rows with missing values in the label column.

 #### Reasoning:

  1. **Supervised Learning Requirement:** The label column is our target variable (ground truth). For supervised learning algorithms, we cannot train on data that lacks the answer we are trying to predict.

  2. **Low Impact (< 5%):** The missing data represents only ~4.7% of the total dataset. Removing these rows leaves us with substantial data for training.

  3. Missing Completely at Random (MCAR): We performed a validity check comparing the behavior profiles (drives, sessions, duration) of users with missing labels versus users with present labels.

  - Result: There was no statistically significant difference in the means of key features between the two groups.
  - Conclusion: The missingness is not structural or biased (e.g., specific to heavy users or iPhone users). It appears to be random.

  4. Data Integrity: Imputing (guessing) the target variable introduces artificial noise and bias into the model's training process. Since the loss of data is minimal and random, dropping is the safest approach to maintain ground truth integrity.
